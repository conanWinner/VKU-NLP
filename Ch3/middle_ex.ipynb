"""
| Ký hiệu     | Ý nghĩa              |
| ----------- | -------------------- |
| `\d` / `\D` | số / không phải số   |
| `\s` / `\S` | khoảng trắng / không |
| `\w` / `\W` | ký tự từ / không     |
| `\b` / `\B` | ranh giới từ / không |
| `\A`        | đầu chuỗi            |
| `\z`, `\Z`  | cuối chuỗi           |
| `\1`, `\2`  | lặp lại group        |
| `\n`, `\t`  | ký tự escape         |

"""

```python
import re
```

```python
test = "I am a student in class in class 2A1 222 . I love to read. I like to read a book with Bill.Bill is a student in class 2A2 "
```

```python
rx_10 = r'\w*\d+\w*'
```

```python
thang = re.findall(rx_10, test)

print(thang)
```

```python
# tách từ
rx_11 = r'\.'
print(re.split(rx_11, test))
```

```python
# stopword 
stop_words = ['I', 'am', 'a', 'in', 'to', 'a', 'with', 'is', 'a', 'in']

words = re.findall(r'\b\w+\b', test.lower())    
filtered = [w for w in words if w not in stop_words]
print(filtered)
```

```python
documents = ["I am a rabbit.", "I live in a hatch.", "I have a fluffy short tail and I have long ears."] 
processed_docs = [doc.lower().replace(".","") for doc in documents]
processed_docs
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#look at the documents list
print("Our corpus: ", processed_docs)

count_vect = CountVectorizer(ngram_range=(2,2))
#Build a BOW representation for the corpus
bow_rep = count_vect.fit_transform(processed_docs)

#Look at the vocabulary mapping
print("Our vocabulary: ", count_vect.vocabulary_)

#see the BOW rep for first 2 documents
print("BoW representation for D1: => ", bow_rep[0].toarray())
print("BoW representation for D3: => ", bow_rep[2].toarray())

# #Get the representation using this vocabulary, for a new text
# temp = count_vect.transform(["dog and dog are friends"])
# print("Bow representation for 'dog and dog are friends':", temp.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vt = TfidfVectorizer()

rep = vt.fit_transform(processed_docs)

print(rep[1].toarray())
```
